**FastScape_test**

Python script to illustrate the use of the *FastScape Interface* in python.

The library file `FastScape.so` should be in the same directory

```python
import numpy as np
import FastScape as Fp
import matplotlib.pyplot as plt
```

This is a simple (low resolution) example where the right-hand side of a rectangular model is an initially 100 m high plateau subjected to erosion, while the left-hand side is kept fixed at base level. The SPL is linear (`n = 1`) but completed by a sediment transport/deposition algoithm with `g = 1`.

Boundary conditions are closed except for the left hand-side (bounday number 4) set to base level.

The model is run for 200 time steps and the results are stored in `.vtk` files where the drainage area is also stored.

```python
Fp.fastscape_init()

nx = 201
ny = 101
nn = nx*ny

Fp.fastscape_set_nx_ny(nx, ny)
Fp.fastscape_setup()

xl = 20.e3
yl = 10.e3
Fp.fastscape_set_xl_yl(xl,yl)

xlist = np.linspace(0, xl, nx)
ylist = np.linspace(0, yl, ny)
X, Y = np.meshgrid(xlist, ylist)
x = X.reshape(nn)
y = Y.reshape(nn)

dt = 2.e3
Fp.fastscape_set_dt(dt)

kf = np.ones(nn)*1.e-4
kfsed = 1.e-4
kd = np.ones(nn)*1.e-2
kdsed = 1.e-2
m = 0.4
n = 1.0
g1 = 1.0
g2 = 1.0
expp = 1.0
Fp.fastscape_set_erosional_parameters(kf, kfsed, m, n, kd, kdsed, g1, g2, expp)

bc = 1
Fp.fastscape_set_bc(bc)

h = np.random.random_sample((nn,))
h = np.where(x > xl/2, h + 1000., h)
Fp.fastscape_init_h(h)

nstep = 200
step = 0
a = np.zeros(nn)
vex = 5.

while step < nstep :
    Fp.fastscape_execute_step()
    step = Fp.fastscape_get_step()
    Fp.fastscape_copy_drainage_area(a)
    Fp.fastscape_vtk(a, vex)

Fp.fastscape_destroy()
```

The drainage area of the last time step is also shown as a contour plot

```python
X = X/1.e3
Y = Y/1.e3
Z = np.reshape(np.log10(a),(ny,nx))
plt.figure()
fig,ax=plt.subplots(figsize=(10,10*yl/xl))
ax.set_aspect(1)
cp = plt.contourf(X, Y, Z)
plt.colorbar(cp)
plt.title('Topography (m)')
plt.xlabel('x (km)')
plt.ylabel('y (km)')
plt.show()
```
